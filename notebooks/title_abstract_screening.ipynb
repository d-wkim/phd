````python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

def title_abstract_screening():
    url = "../data/test.csv"
    df = pd.read_csv(url, encoding = "utf-8")
    df["id"] = pd.DataFrame({
        "id": range(1,len(df))})
    df = df.sort_values(by = "id")
    a = widgets.IntText(description = "ID ", min = 1, max = len(df)-1, layout = {"width": "50%"})
    # figure out later how to include only ids that exist
    a3 = widgets.HTML(value = "/" + str(len(df)))
    #link = widgets.jslink((a,"value"),(a1,"value"))
    b = widgets.HTML(value = "", description = "study ", layout = {"width": "50%"}) # study
    c = widgets.HTML(description = "subgroup", layout = {"width": "50%"})
    d = widgets.HTML(value = "", description = "author(s) ", layout = {"width": "90%"}) # authors
    e = widgets.HTML(value = "", description = "title ", layout = {"width": "90%"}) # title
    f = widgets.HTML(value = "", description = "abstract ", layout = {"width": "91.5%", "height": "350px"}) # abstract
    g = widgets.HTML(value = "", description = "year ", layout = {"width": "90%"}) # year
    h = widgets.HTML(value = "", description = "journal ", layout = {"width": "90%"}) # journal
    i = widgets.Dropdown(options = ["PubMed", "Embase", "Web of Science"], value = "PubMed", description = "source", layout = {"width": "50%"})
    j = widgets.HTML(value = "", description = "DOI ", layout = {"width": "90%"}) # doi
    k = widgets.HTML(value = "", description = "URL ", layout = {"width": "90%"}) # url
    result = widgets.Text()
    
    def update(change):
        study = df.loc[df["id"] == a.value, "study"]
        subgroup =  df.loc[df["id"] == a.value, "subgroup"]
        authors =  df.loc[df["id"] == a.value, "authors"]
        title =  df.loc[df["id"] == a.value, "title"]
        abstract =  df.loc[df["id"] == a.value, "abstract"]
        year =  df.loc[df["id"] == a.value, "year"]
        journal =  df.loc[df["id"] == a.value, "journal"]
        source =  df.loc[df["id"] == a.value, "source"]
        doi =  df.loc[df["id"] == a.value, "doi"]
        url =  df.loc[df["id"] == a.value, "doi_url"]
        result =  df.loc[df["id"] == a.value, "title_abstract_screening"]
  
    
        if a.value > 0:
           b.value = str(study.iloc[0])
           c.value = str(subgroup.iloc[0])
           d.value = str(authors.iloc[0])
           e.value = str(title.iloc[0])
           f.value = str(abstract.iloc[0])
           g.value = str(year.iloc[0])
           h.value = str(journal.iloc[0])
           j.value = str(doi.iloc[0])
           k.value = str(url.iloc[0])
           result.value = str(title_abstract_screening.iloc[0])
    
        else:
            b.value = ""
            c.value = ""
            d.value = ""
            e.value = ""
            f.value = ""
            g.value = ""
            h.value = ""
            j.value = ""
            k.value = ""
            result.value = ""
    
    a.observe(update, names = "value") # observe widget 'a' (i.e., ID column) for any changes, and if there is, then display all of the widgets corresponding to the value of 'a'.
    
    #save = widgets.Button(value = "Save", description = "Save", button_style = "info")
    def save_click(button):
        # save dataset from clicking only yes, maybe, no.
        import os
        os.makedirs("../data", exist_ok = True)        
        df.to_csv("../data/title_abstract_screening.csv", encoding = "utf-8")
        
    
    
    yes = widgets.Button(value = "Yes", description = "Yes", button_style = "success")
    def yes_click(button):
        df.loc[df["id"] == a.value, "title_abstract_screening"] = "yes"
        yes.on_click(save_click)
    
    
    maybe = widgets.Button(value = "Maybe", description = "Maybe", button_style = "warning")
    def maybe_click(button):
        df.loc[df["id"] == a.value, "title_abstract_screening"] = "maybe"
        maybe.on_click(save_click)
        
    no = widgets.Button(value = "No", description = "No", button_style = "danger")
    def no_click(button):
        df.loc[df["id"] == a.value, "title_abstract_screening"] = "no"
        no.on_click(save_click)
    
    yes.on_click(yes_click)
    maybe.on_click(maybe_click)
    no.on_click(no_click)

    out = widgets.Output()

    buttons = widgets.VBox(children = [yes, maybe, no])
    A = widgets.HBox(children = [a, a3])
    Abcd = widgets.VBox(children = [A, b, c, d])
    
    title = widgets.HTML(value = "<h1 align='center' >Title and abstract screening form<h1>")
    
    display(title)
    row = widgets.HBox(children = [Abcd, buttons])
    display(row)
    w_e = display(e, out)
    w_f = display(f, out)
    w_g = display(g, out)
    w_h = display(h, out)
    w_i = display(i, out)
    w_j = display(j, out)
    w_k = display(k, out)
    
    form = widgets.HBox(children = buttons)
    display(form, out)

title_abstract_screening()

outcomes = widgets.SelectMultiple(options = ["ikdc_subjective",
    "lysholm",
    "tegner",
    "instrumental_laxity",
    "pivot_shift",
    "lachman",
    "graft_rupture"
])

display(outcomes)
    
````
